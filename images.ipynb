## Flip

```python
def flip(img_name):
    
    # open original image and extract pixel map
    originalImg = Image.open(img_name).convert("L") # open and convert to greyscale
    pixelMap = originalImg.load()
    
    # get width and height
    width, height = originalImg.size # width = 450, height = 300
    
    # initialize new image
    newImg = Image.new('L', (width, height))
    pixelMapNew = newImg.load()
    
    # flip new image
    for x in range(width):
        for y in range(height):
            pixelMapNew[x,y] = pixelMap[(width-x-1), y] 
            
    display(newImg)
    
flip('eye.png')

# notes: requires only a flip on the horizontal axis (x-axis), 
# and requires an offset of 1 to stay within index bounds.
# Personally, the trickiest part was understanding the relationship
# between the extracted pixel map and resultant image. I didn't know
# that the extracted pixel map would automatically update the parent
# image. Python wizardry!
```
